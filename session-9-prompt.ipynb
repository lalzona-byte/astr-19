```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
x = np.random.uniform(0,1,1000) 

plt.hist(x, bins=100, edgecolor='black') 

plt.xlabel('x')
plt.ylabel('data per bin') 
plt.title('Prompt 9 Histogram')
plt.savefig("session-9-histo.pdf",bbox_inches="tight", facecolor="white")

```
