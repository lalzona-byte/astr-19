```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
x = np.random.standard_gamma(shape=3.0, size=1000)

plt.hist(x, bins=100, edgecolor='black')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Prompt 10 Histogram')

plt.savefig("session-10-hist.pdf", bbox_inches="tight", facecolor="white") 
```
